```python
# First thing we need to install wget in jupyter notebook
# !pip install wget

import pandas as pd
import numpy as np
import wget 
```

### **Using wget to download all the files**

```python
# Download csv files using wget

urls = ['https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv',
        'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv',
         'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv']

for url in urls:
    wget.download(url)


```

### **Read all the files**

```python
df_confirmed = pd.read_csv('time_series_covid19_confirmed_global.csv')

df_deaths = pd.read_csv('time_series_covid19_deaths_global.csv')

df_recovered = pd.read_csv('time_series_covid19_recovered_global.csv')
```

```python
#  Set option to see all the rows

pd.set_option('display.max_rows', None)

#  Set option to see all the columns

pd.set_option('display.max_columns', None)
```

### Let's have a quick look at the three tables, we see that all dates are in pivot form, which becomes difficult to 

### combine all of them into one.

```python
df_confirmed.head()
```

#####  All three tables have exactly 1020 columns in total but less rows in df_recovered

#####  That is because Canada's recovered data was recorded by Country/Region-wise rather than Province/State-wise in the 

#####  df_recovered table

```python
#  showing the shapes of all three tables

print(df_confirmed.shape)

print(df_deaths.shape)

print(df_recovered.shape)

df_confirmed_count = df_confirmed[['Province/State', 'Country/Region', 'Lat', 'Long']].count()
print(df_confirmed_count)


df_deaths_count = df_deaths[['Province/State', 'Country/Region', 'Lat', 'Long']].count()
print(df_deaths_count)


df_recovered_count = df_recovered[['Province/State', 'Country/Region', 'Lat', 'Long']].count()
print(df_recovered_count)
```

##### Since all the dates are in pivot form, which is not ideal to perform merge()

##### We have to use melt() to unpivot DataFrames from current wide format into long format. 

```python
df_confirmed_unpivot = df_confirmed.melt( id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                       var_name = 'Date', 
                                       value_name = 'Confirmed'
                                         )

df_deaths_unpivot = df_deaths.melt( id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                       var_name = 'Date', 
                                       value_name = 'Deaths'
                                   )

df_recovered_unpivot = df_recovered.melt( id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                       var_name = 'Date', 
                                       value_name = 'Recovered'
                                         )
```

```python
df_confirmed_unpivot.head()
```

```python
df_recovered_unpivot.head()
```

##### We are going to use merge() to combine all three tables 

```python
df_combined = pd.merge(df_confirmed_unpivot, 
                       df_deaths_unpivot, 
                       how = 'left', 
                       on = ['Province/State', 'Country/Region', 'Lat', 'Long', 'Date']
                      )
                      
df_combined = pd.merge(df_combined, 
                       df_recovered_unpivot, 
                       how = 'left', 
                       on = ['Province/State', 'Country/Region', 'Lat', 'Long', 'Date']
                      )
                    
```

##### We have to remove Canada from the table since Canada's revovered data was counted as Country-wise

```python
df_combined = df_combined.loc[df_combined['Country/Region'] != 'Canada']
```

```python
df_combined.head(2000)
```

### Next, we are ready to perform data cleaning

```python
df_combined.info()
```

##### Change dtypes in Date column

```python
df_combined['Date'] = pd.to_datetime(df_combined['Date'])
```

##### We see that there are Null values in 'Province/State', 'Lat', 'Long' and 'Recovered'

##### There are tons of Null values in 'Province/State', which make sense. Since the data of certain countries were 

##### reported through Country/Region-wise

```python
df_combined.isnull().sum()
```

##### Thoses Null values in 'Lat' and 'Long' are coming from 'Repatriated Travellers' and 'Unknown' in 'Province/State' Column

##### We need to get rid of the rows containing the Null values in 'Lat' and 'Long'

```python
#  Drop the rows containing 'Repatriated Travellers' and 'Unknown'

df_combined = df_combined.loc[(df_combined['Province/State'] != 'Repatriated Travellers') 
                              & 
                              (df_combined['Province/State'] != 'Unknown')
                             ]
```

##### We found that 'Lat' and 'Long' also have zero values rather than Null

##### We have to filter out those rows and drop them

```python
# First thing we have to do is to change dtype of 'Lat' and 'Long' into int

df_combined_temp = df_combined[['Lat', 'Long']].astype('int')
```

```python
# Then we can filter out those rows which contain zero values in 'Lat' and 'Long'

# We found that 'Diamond Princess' and 'MS Zaandam' have mismatch data in the table

df_unuseful = df_combined.loc[(df_combined_temp['Lat'] == 0) 
                              & 
                              (df_combined_temp['Long'] == 0)
                             ]
df_unuseful.head()
```

```python
# We are going to drop them

df_combined = df_combined.loc[(df_combined['Country/Region'] != 'Diamond Princess') 
                              &
                              (df_combined['Country/Region'] != 'MS Zaandam')]
```

##### We are going to replace Nulls with zeros in 'Recovered' column

```python
df_combined['Recovered'] = df_combined['Recovered'].fillna(0)
```

##### We are going to add a new column called 'Active', which is calculated by 'Confirmed' - 'Deaths' - 'Recovered'

##### Note: Confirmed, Deaths, Recovered and Active are cumulative data.

```python
df_combined['Active'] = df_combined['Confirmed'] - df_combined['Deaths'] - df_combined['Recovered']
```

```python
df_combined.head()
```

##### Let's group data into 'Country/Region' and Date' by calculating the sum of 'Confirmed', 'Deaths', 'Recovered' and 'Active' 

```python
df_combined_groupby = df_combined.groupby(['Country/Region', 'Date'])[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().reset_index()
```

```python
df_combined_groupby.head()
```

##### Let's create new columns : 'New cases', 'New deaths' and 'New recovered'

##### These columns are calculated as the current row - previous row since df_combined is cumulative

##### Note: New cases, New deaths and New Recovered are day wise data

```python
df_new_groupby = df_combined.groupby(['Country/Region', 'Date'])[['Confirmed', 'Deaths', 'Recovered']].sum().diff().reset_index()


df_new_groupby.columns = ['Country/Region', 'Date', 'New cases', 'New deaths', 'New recovered']


df_new_groupby = df_new_groupby.fillna(0)
```

##### Let's merge two new tables

```python
df_new_combined = pd.merge(df_combined_groupby, df_new_groupby, how = 'inner', on = ['Country/Region', 'Date'])
```

```python
df_new_combined.head()
```

##### Finally, let's replace all negative values in 'New cases',  'New deaths'	and 'New recovered' with 0

```python
df_new_combined['New cases'] = df_new_combined[df_new_combined['New cases'] < 0] = 0
df_new_combined['New deaths'] = df_new_combined[df_new_combined['New deaths'] < 0] = 0
df_new_combined['New recovered'] = df_new_combined[df_new_combined['New recovered'] < 0] = 0
```

##### Let's save this file 

```python
df_new_combined.to_csv('Covid-19 Data Cleaning Completed.csv')
```
